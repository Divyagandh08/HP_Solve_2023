```r
a=table(customer_data$reviewDislikes)
barplot(a,main="Using BarPlot to display reviewDislikes Comparision",
       ylab="Count",
       xlab="Dislikes",
       col=rainbow(1),
       legend=rownames(a))
```

```r
a=table(customer_data$reviewLikes)
barplot(a,main="Using BarPlot to display reviewlikes Comparision",
       ylab="Count",
       xlab="Likes",
       col=blues9,
       legend=rownames(a))
```

```r
install.packages("plotrix")
```

```r
pct=round(a/sum(a)*100)
lbs=paste(c("$reviewLikes","$reviewDislikes")," ",pct,"%",sep=" ")
library(plotrix)
pie3D(a,labels=lbs,main="Pie Chart Depicting Ratio of reviewLikes, reviewDislikes")
```
